```python
# if in general 66% students like python find the probability that exactly 3 out of 5 students will like python- binomial distribution probelem
from scipy.stats import binom
# Given values
N = 5 # Total students
K = 3 # Students who like python
P = 0.66 # Probability that student like python
# Calculate binomial probability
Probability = binom.pmf(K,N,P)
print(Probability)
```

##### Hospital records show that of patients suffering from a particular disease, 75% die of it. What is the probability that out of 6 patients suffering from the disease, four will recover? - binomial distribution problem

```python
N = 6 # Total patients
K = 2 # Patients who die
P = 0.75 # Probability that patient die
# Calculate binomial probability
Probability = binom.pmf(K,N,P)
print(Probability)
```

##### A blindfold marksman finds that on the average, he hits the bull's eye 4 times in 5 shots. What is the probability that in the next 4 shots, he will hit the bull's eye a) more than 2 times b) at least 3 misses

```python
N = 4 # Total shots
P = 4/5 # Probability that he hits bull's eye
# a) more than 2 times
K = 3 # Hits more than 2 times
Probability_more_than_2 = binom.sf(K-1,N,P) # sf is 1-cdf
print("Probability of hitting more than 2 times:", Probability_more_than_2)
# b) at least 3 misses
K = 1 # Hits at most 1 time (at least 3 misses)
Probability_at_least_3_misses = binom.cdf(K,N,P)
print("Probability of at least 3 misses:", Probability_at_least_3_misses)
```

##### 85% borrowers repay their loans on time. What is the probability that out of 10 borrowers, a) more than 7 will repay b) at least 3 default

```python
N = 10
P = 0.85 # Probability that borrower repays loan on time
# a) more than 7 will repay
K = 7
Probability_more_than_7 = binom.sf(K,N,P)
print("Probability of more than 7 repaying loans on time:", Probability_more_than_7)
# b) at least 3 default
K = 7 # At most 7 will repay (at least 3 default)
Probability_at_least_3_default = binom.cdf(K,N,P)
print("Probability of at least 3 defaulting:", Probability_at_least_3_default)
```
