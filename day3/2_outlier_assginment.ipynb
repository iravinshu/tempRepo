```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df['age'], orient='h')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(x=df['age'].dropna(), vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
age_data = df['age']

Q1 = age_data.quantile(0.25)
Q3 = age_data.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = age_data[(age_data < lower_bound) | (age_data > upper_bound)]
outlier_count = len(outliers)
outlier_count
```

```python
df = pd.read_csv('../dataset/tips.csv', sep = ',')
df
```

```python
q1 = np.percentile(df['total_bill'], 25) # 25th percentile
q3 = np.percentile(df['total_bill'], 75) # 75th percentile
iqr = q3 - q1
lower_fence = q1 - 1.5*iqr
upper_fence = q3 + 1.5*iqr
```

```python
print(f"Q1: {q1}\nQ3: {q3}\nQ2: {np.median(df['total_bill'])}\nLower Fence: {lower_fence}\nUpper Fence: {upper_fence}\nIQR: {iqr}")
```

```python
outliers = df[(df['total_bill'] < lower_fence) | (df['total_bill'] > upper_fence)]['total_bill']
```

```python
outliers.sort_values()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df['total_bill'], orient='h')
plt.title('Box Plot of Total bill')
plt.xlabel('Total Bill')
plt.show()
```
