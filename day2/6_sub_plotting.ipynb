```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
```

```python
plt.figure(figsize=(20,5))
avg_tip_day = df.groupby('day')['tip'].mean()
plt.subplot(1,2, 1)
plt.bar(avg_tip_day.index, avg_tip_day.values, color=['lightblue', 'lightgreen', 'pink', 'orange'])
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
smoker_counts = df['smoker'].value_counts()
plt.subplot(1,2, 2)
plt.pie(smoker_counts.values, labels=smoker_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Smoker vs Non-Smoker')
# plt.tight_layout()
plt.show()
```

### Subplot 2 horizontal bar of average tip by gender

```python
#=========2. obeject oriented syntax
fig,axes = plt.subplots(1,2,figsize=(12,5))
#subplot 1: line plot of total_bill ober the first 20 rows
axes[0].plot(df['total_bill'].iloc[:20],marker='o',linestyle='-',color='purple')
axes[0].set_title('Total Bill (First 20 records)')
axes[0].set_xlabel('Record Index')
axes[0].set_ylabel('Total Bill')
# subplot 2: horizontal bar of average tip by gender
average_tip_by_gender = df.groupby('gender')['tip'].mean()
axes[1].barh(average_tip_by_gender.index,average_tip_by_gender.values,color='pink')
axes[1].set_title('Avg. Tip by gender')
axes[1].set_xlabel('Average Tip')
# axes[1].set_ylabel('Total Bill')
plt.show()
```
