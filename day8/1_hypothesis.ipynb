#### It is estimated that the average marks obtained by students in some entrance exam is 76 out of 100. A random sample of 100 students were taken and their average marks were 78 with a standard deviation of 2.
1) State your null and alternate hypothesis.
2) Decide the alpha value
3) Find the test statistic
4) Compare the test statistic with critical value
5) Compare your p-value with alpha value
6) Write your conclusion about null hypothesis

```python
s = 2
n = 100
x_bar = 78
mu = 76
zscore = (x_bar - mu) / (s / (n ** 0.5))
print(f"Z score: {zscore:.2f}")
print(f"Since the calculated z score {zscore:.2f} is greater than the critical value 1.96, we reject the null hypothesis.")
```

```python
alpha = 0.05
from scipy import stats
p_value = 2 * (1 - stats.norm.cdf(abs(zscore)))
print(f"P value: {p_value:.4f}")
```

```python
print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis.")
```

```python
print("Rejection of null hypothesis means that there is enough evidence to support the claim that the average marks obtained by students in the entrance exam is significantly different from 76.")
```

#### A pharma company claims that the average effectiveness of the medicine can be seen in 8 weeks. A random sample of 25 patient was taken and the seen in 8.5. Do same as above with alpha value of 0.01, 0.05, 0.1 separately.

```python
s = 0.25
n = 25
x_bar = 8.5
mu = 8
zscore = (x_bar - mu) / (s / (n ** 0.5))
print(f"Z score: {zscore:.2f}")
print(f"Since the calculated z score {zscore:.2f} is greater than the critical value 2.576, we reject the null hypothesis at alpha 0.01.")
```

```python
alpha = 0.01
from scipy import stats
p_value = 2 * (1 - stats.norm.cdf(abs(zscore)))
print(f"P value: {p_value:.4f}")
print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis at alpha 0.01.")
```

```python
alpha = 0.05
print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis at alpha 0.05.")
```

```python
alpha = 0.1
print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis at alpha 0.1.")
```
