### Find average age of female and male passengers

```python
avg_male_age = df[df['gender'] == 'male']['age'].mean()
avg_female_age = df[df['gender'] == 'female']['age'].mean()
print("Average Male Age:", avg_male_age)
print("Average Female Age:", avg_female_age)
```

#### 6. Separate passengers who survived (survived column contains 0) and who did not survive (survived column contains 1) into two separate dataframes.


```python
df_survived = df[df['survived'] == 1]
df_not_survived = df[df['survived'] == 0]
```

#### 7. For the above two dataframes, find number of females and males.


```python
male_count_survived = df_survived[df_survived['gender'] == 'male'].shape[0]
female_count_survived = df_survived[df_survived['gender'] == 'female'].shape[0]
male_count_not_survived = df_not_survived[df_not_survived['gender'] == 'male'].shape[0]
female_count_not_survived = df_not_survived[df_not_survived['gender'] == 'female'].shape[0]
print("Survived - Males:", male_count_survived, "Females:", female_count_survived)
print("Not Survived - Males:", male_count_not_survived, "Females:", female_count_not_survived)
```
