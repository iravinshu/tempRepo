```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Graph

### MatplotLib

```python
df = pd.read_csv('tips.csv')
df
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'], bins=10, color='skyblue', edgecolor='black')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

### Seaborn

```python
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], bins=10, kde=True, color='darkgreen')
plt.title('Total Bill Distribution with KDE')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
avg_tip_gender = df.groupby('gender')['tip'].mean()
plt.bar(avg_tip_gender.index, avg_tip_gender.values, color=['pink', 'lightblue'])
plt.title('Average Tip by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure(figsize=(10,6))
avg_tip_gender = df.groupby('gender')['tip'].mean().reset_index()
sns.barplot(avg_tip_gender, x='gender', y='tip', palette=['pink', 'lightblue'], hue='tip')
plt.title('Average Tip by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['total_bill'], df['tip'], color='orange', edgecolor='black')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='total_bill', y='tip', color='red')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='gender', y='tip', palette='Set2', hue='gender')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([df[df['gender'] == gender]['tip'] for gender in df['gender'].unique()],
            labels=df['gender'].unique())
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='smoker', palette='pastel', hue='smoker')
plt.title('Smoker Distribution')
plt.ylabel('Count')
plt.show()
```
