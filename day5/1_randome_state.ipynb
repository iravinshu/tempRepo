```python
import pandas as pd
```

```python
df_2015 = pd.read_csv('../dataset/marathon_results_2015.csv', sep = ',')
df_2016 = pd.read_csv('../dataset/marathon_results_2016.csv', sep = ',')
df_2017 = pd.read_csv('../dataset/marathon_results_2017.csv', sep = ',')
```

```python
df = pd.concat([df_2015, df_2016, df_2017])
df
```

```python
original_count = len(df)
df_random_state = df.sample(frac=0.1, random_state=42)
shuffled_count = len(df_random_state)
```

```python
print(original_count, shuffled_count)
```

```python
total_males = len(df[df['M/F'] == 'M'])
total_females = len(df[df['M/F'] == 'F'])
```

```python
sample_size = int(0.1 * original_count)
sample_males = int(0.1 * total_males)
sample_females = int(0.1 * total_females)
print("Sample Size:", sample_size)
print("Sample Males:", sample_males)
print("Sample Females:", sample_females)
```

```python
male_sample = df[df['M/F'] == 'M'].sample(n=sample_males, random_state=42)
female_sample = df[df['M/F'] == 'F'].sample(n=sample_females, random_state=42)
print(len(male_sample), len(female_sample))
```

```python
sample_df = pd.concat([male_sample, female_sample])
sample_row_count = len(sample_df)
sample_df.head()
```

```python
unique_countries = df['Country'].unique()
print("Unique Countries in Original DataFrame:", unique_countries)
```

```python
num_countries = len(unique_countries)
```

```python
import numpy as np
np.random.seed(42)
# choice function in numpy generates a random sample from an array
random_countries = np.random.choice(unique_countries, size=50, replace=False)
print("Randomly Selected Countries:", random_countries)
```

```python
sample_df = df[df['Country'].isin(random_countries)]
sample_row_count = len(sample_df)
```
