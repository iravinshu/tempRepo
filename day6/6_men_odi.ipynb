```python
import pandas as pd
from scipy.stats import binom
men_20 = pd.read_csv('../dataset/Men ODI Team Match Results - 20th Century.csv', sep = ',')
men_20
```

```python
men_21 = pd.read_csv('../dataset/Men ODI Team Match Results - 21st Century.csv', sep = ',')
men_21
```

```python
df = pd.concat([men_20, men_21])
df
```

##### find the historical win rate of india
##### now india is scheduled to play 60 odi matches over the next 3 years

```python
df['team1'] = df['Match'].str.split(' v ').str[0]
df['team2'] = df['Match'].str.split(' v ').str[1]
```

```python
total_india_matches = df[(df['team1'] == 'India') | (df['team2'] == 'India')]
total_india_matches
```

```python
total_india_won = total_india_matches[((total_india_matches['Country'] == 'India') & (total_india_matches['Result'] == 'Won')) | ((total_india_matches['Country'] != 'India') & (total_india_matches['Result'] == 'Lost'))]
total_india_won
```

```python
india_win_rate = len(total_india_won) / len(total_india_matches)
print('{:.2f}'.format(india_win_rate * 100) + "%")
```

##### a) what is probability that india wins exactly 35 matches out of 60

```python
N = 60
K = 35
p = india_win_rate
prob = binom.pmf(K, N, p)
print('{:.2f}'.format(prob * 100) + "%")
```

##### b) what is probability that india wins more than 40 matches

```python
N = 60
K = 40
p = india_win_rate
prob = 1 - binom.cdf(K, N, p)
print('{:.2f}'.format(prob * 100) + "%")
```

##### c) what is probability that india wins at least 45 matches

```python
N = 60
K = 45
p = india_win_rate
prob = 1 - binom.cdf(K-1, N, p)
print('{:.2f}'.format(prob * 100) + "%")
```

##### d) what is the most likely number of wins India will have out of 60 matches

```python
N = 60
p = india_win_rate
most_likely_wins = N * p
str(round(most_likely_wins)) + " matches"
```
