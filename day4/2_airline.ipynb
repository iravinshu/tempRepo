```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/airline_passenger_satisfaction.csv', sep = ',')
df
```

### Do a skewness analysis and plotting for the 'On-board Service' column of 'airline_passenger_satisfaction.csv' dataset. Also find CoV for first time versus returning customers for the 'Ease of Online Booking' column.

```python
onboard_service = df['On-board Service']
mean_onboard_service = onboard_service.mean()
median_onboard_service = onboard_service.median()
skew_onboard_service = onboard_service.skew()
kurtosis_onboard_service = onboard_service.kurt()
```

```python
print("On-board Service Statistics:")
print(f"Mean: {mean_onboard_service}")
print(f"Median: {median_onboard_service}")
print(f"Skewness: {skew_onboard_service}")
print(f"Kurtosis: {kurtosis_onboard_service}")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(onboard_service, bins=30, kde=True, color='skyblue')
plt.axvline(mean_onboard_service, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_onboard_service:.2f}')
plt.axvline(median_onboard_service, color='green', linestyle='--', linewidth=2, label=f'Median: {median_onboard_service:.2f}')
plt.title('Distribution of On-board Service\nSkewness: {:.2f}, Kurtosis: {:.2f}'.format(skew_onboard_service, kurtosis_onboard_service))
plt.xlabel('On-board Service Rating')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=onboard_service, color='lightgreen')
plt.title('Box Plot of On-board Service')
plt.xlabel('On-board Service Rating')
plt.tight_layout()
plt.show()
```
