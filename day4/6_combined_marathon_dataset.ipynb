#### 6) Combine the 3 marathon datasets (Hint: df = pd.concat([df_2015, df_2016, df_2017]) ). Then do a correlation analysis on 'Official time' (i.e. time taken to complete the race) and 'pace' (i.e. average pace per minute in km) columns. Draw scatter plot and heatmap.

```python
import pandas as pd
marathon_results_2015 = pd.read_csv('../dataset/marathon_results_2015.csv', sep = ',')
marathon_results_2015
```

```python
import pandas as pd
marathon_results_2016 = pd.read_csv('../dataset/marathon_results_2016.csv', sep = ',')
marathon_results_2016
```

```python
import pandas as pd
marathon_results_2017 = pd.read_csv('../dataset/marathon_results_2017.csv', sep = ',')
marathon_results_2017
```

```python
df = pd.concat([marathon_results_2015, marathon_results_2016, marathon_results_2017])
df
```

```python
# convert Official Time to total minutes in integer format
def time_to_minutes(time_str):
    h, m, s = map(int, time_str.split(':'))
    return h * 60 + m + s / 60
df['Official Time'] = df['Official Time'].apply(time_to_minutes)
df['Pace'] = df['Pace'].apply(time_to_minutes)
```

```python
corr_overall = df[['Official Time','Pace']].corr()
cov_overall = df[['Official Time','Pace']].cov()
```

```python
print(corr_overall, "\n\n")
print(cov_overall, "\n\n")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Official Time'],df['Pace'],color='green',alpha=0.5)
plt.title(f'overall correlation: {corr_overall.loc["Official Time","Pace"]:.2f}')
plt.xlabel('Official Time')
plt.ylabel('Pace')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#heatmap
import seaborn as sns
plt.figure(figsize=(6,4))
sns.heatmap(corr_overall,annot=True,fmt='.2f',cmap='coolwarm')
plt.title('overall correlation heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
