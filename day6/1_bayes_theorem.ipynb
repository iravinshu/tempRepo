```python
import pandas as pd
titanic = pd.read_csv('../dataset/titanic.csv', sep = ',')
titanic
```

```python
total_passengers = len(titanic)
survived_count = len(titanic[titanic['survived'] == 1])
female_count = len(titanic[titanic['gender'] == 'female'])
P_survived = survived_count / total_passengers
P_female = female_count / total_passengers
P_female_and_survived = len(titanic[(titanic['survived'] == 1) & (titanic['gender'] == 'female')]) / total_passengers
```

#### Apply Bayes Theorem to find P(survived|female)

```python
P_survived_given_female = (P_female_and_survived / P_female) * P_survived
P_survived_given_female
```
