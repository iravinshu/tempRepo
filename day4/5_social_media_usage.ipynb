```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/social_media_usage.csv', sep = ',')
df
```

#### Do a correlation analysis between Social media usage on facebook and instagram users. Columns to be used: daily minutes per day and posts per day.

```python
df_facebook = df[df['App']=='Facebook']
df_instagram = df[df['App']=='Instagram']
```

```python
df_facebook
```

```python
corr_overall = df[['Daily_Minutes_Spent','Posts_Per_Day']].corr()
corr_facebook = df_facebook[['Daily_Minutes_Spent','Posts_Per_Day']].corr()
corr_instagram = df_instagram[['Daily_Minutes_Spent','Posts_Per_Day']].corr()
```

```python
cov_overall = df[['Daily_Minutes_Spent','Posts_Per_Day']].cov()
cov_facebook = df_facebook[['Daily_Minutes_Spent','Posts_Per_Day']].cov()
cov_instagram = df_instagram[['Daily_Minutes_Spent','Posts_Per_Day']].cov()
```

```python
print(corr_overall,corr_facebook,corr_instagram)
print(cov_overall,cov_facebook,cov_instagram)
```

```python
fig,axes = plt.subplots(1,3,figsize=(18,6))
axes[0].scatter(df['Daily_Minutes_Spent'],df['Posts_Per_Day'],color='green',alpha=0.5)
axes[0].set_title(f'overall correlation: {corr_overall.loc["Daily_Minutes_Spent","Posts_Per_Day"]:.2f}')
axes[0].set_xlabel('Daily_Minutes_Spent')
axes[0].set_ylabel('Posts_Per_Day')
axes[0].grid(True)
#male data plot
axes[1].scatter(df_facebook['Daily_Minutes_Spent'],df_facebook['Posts_Per_Day'],color='green',alpha=0.5)
axes[1].set_title(f'male correlation: {corr_facebook.loc["Daily_Minutes_Spent","Posts_Per_Day"]:.2f}')
axes[1].set_xlabel('Daily_Minutes_Spent')
axes[1].set_ylabel('Posts_Per_Day')
axes[1].grid(True)
#female data plots
axes[2].scatter(df_instagram['Daily_Minutes_Spent'],df_instagram['Posts_Per_Day'],color='green',alpha=0.5)
axes[2].set_title(f'female correlation: {corr_instagram.loc["Daily_Minutes_Spent","Posts_Per_Day"]:.2f}')
axes[2].set_xlabel('Daily_Minutes_Spent')
axes[2].set_ylabel('Posts_Per_Day')
axes[2].grid(True)
plt.tight_layout()
plt.show()
```
