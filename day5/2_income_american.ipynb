```python
import pandas as pd
df = pd.read_csv('../dataset/income.csv', sep = ',')
df
```

```python
white_df = df[df['race']=="White"]
black_df = df[df['race']=="Black"]
other_df = df[df['race']=="Other"]
asian_df = df[df['race']=="Asian-Pac-Islander"]
amer_df = df[df['race']=="Amer-Indian-Eskimo"]
```

```python
sample_white_df = white_df.sample(frac=0.1, random_state=42)
sample_black_df = black_df.sample(frac=0.1, random_state=42)
sample_other_df = other_df.sample(frac=0.1, random_state=42)
sample_asian_df = asian_df.sample(frac=0.1, random_state=42)
sample_amer_df = amer_df.sample(frac=0.1, random_state=42)
sample_df = pd.concat([sample_white_df, sample_black_df, sample_other_df, sample_asian_df, sample_amer_df])
print(len(sample_df), len(df))
```
