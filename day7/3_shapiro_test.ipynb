```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
data = pd.read_csv('../dataset/tips.csv')
data
```

```python
total_bill_series = data['total_bill']
```

```python
shapiro_test = stats.shapiro(total_bill_series)
shapiro_test
```

```python
print("test statistic (W) :", shapiro_test.statistic)
```

```python
print("p-value :", shapiro_test.pvalue)
```

```python
alpha = 0.05 # common significance level
if shapiro_test.pvalue > alpha:
    print("Fail to reject the null hypothesis: Data is normally distributed")
else:
    print("Reject the null hypothesis: Data is not normally distributed")
```

#### QQ Plot

```python
stats.probplot(total_bill_series, dist='norm', plot=plt)
plt.title('QQ Plot of Total Bill')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Central Limit Theorem Demonstration

```python
tests = []
for i in range(100):
    mean = diabetes['Glucose'].sample(n=50, random_state=i).mean()
    tests.append(mean)
print(stats.shapiro(tests).pvalue > 0.05)
```

```python
gulucose_shapiro = stats.shapiro(diabetes['Glucose'].dropna())
print(gulucose_shapiro)
```

```python
stats.probplot(diabetes['Glucose'].dropna(), dist='norm', plot=plt)
plt.title('QQ Plot of Glucose')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
sample_means = []
n_sample = 100
sample_size = 768
for i in range(n_sample):
    sample = diabetes['Glucose'].dropna().sample(n=sample_size, random_state=i)
    sample_means.append(sample.mean())
print(sample_means)
```

```python
sample_means_shapiro = stats.shapiro(sample_means)
print(sample_means_shapiro)
if sample_means_shapiro.pvalue < 0.05:
    print("Reject the null hypothesis: Sample means are not normally distributed")
else:
    print("Fail to reject the null hypothesis: Sample means are normally distributed")
```
