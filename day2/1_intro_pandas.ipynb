```python
import pandas as pd
import numpy as np
```

```python
a = np.array([10, 20, 30])
b = np.array([1,2,3])
print(a+b)
s1 = pd.Series([10,20,30], index=['alice', 'bob', 'charlie'])
s2 = pd.Series([1,2,3], index=['bob', 'alice', 'charlie'])
print(s1+s2)
```

```python
s1 = pd.Series([10,20,30,40])
s2 = pd.Series([1,2,3,4], index=['bob', 'alice', 'charlie', 'dev'])
```

```python
print(s1+s2)
```

```python
marks = [67, 36, 81]
names = ['Alice', 'Bob', 'Charlie']

students = pd.Series(marks)
print(students)
```

```python
students = pd.Series(marks, index=names)
print(students)
```

```python
random_ages = np.random.randint(0,100,4)
print(random_ages)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Dev'] # Length of both array matters else pandas will give error
```

```python
people_array = pd.Series(random_ages, index=names)
print(people_array)
```
