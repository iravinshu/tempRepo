```python
import pandas as pd
titanic = pd.read_csv('../dataset/titanic.csv', sep = ',')
titanic
```

#### Probablity that passenger dies

```python
probability_dies = len(titanic[titanic['survived'] == 0]) / len(titanic)
probability_dies
```

#### Probability that passenger survives and male

```python
probability_survives_m = len(titanic[(titanic['survived'] == 1) & (titanic['gender'] == 'male')]) / len(titanic)
probability_survives_m
```

#### Probability that passenger survives given that she was female

```python
probability_survives_f = len(titanic[(titanic['survived'] == 1) & (titanic['gender'] == 'female')]) / len(titanic[titanic['gender'] == 'female'])
probability_survives_f
```
