```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
commuter_times = [16, 8, 35, 17, 13, 15, 15, 5, 16, 25, 20, 12, 10]
data_array = np.array(commuter_times)
```

```python
q1 = np.percentile(data_array, 25) # 25th percentile
q3 = np.percentile(data_array, 75) # 75th percentile
iqr = q3 - q1
lower_fence = q1 - 1.5*iqr
upper_fence = q3 + 1.5*iqr
```

```python
print(f"Q1: {q1}\nQ3: {q3}\nQ2: {np.median(data_array)}\nLower Fence: {lower_fence}\nUpper Fence: {upper_fence}\nIQR: {iqr}")
```

```python
outliers = data_array[(data_array < lower_fence) | (data_array > upper_fence)]
```

```python
print(outliers)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=commuter_times, orient='h')
plt.text(x=20, y=5, s="Highest tip", fontsize=12, color='red')
plt.title('Box Plot of Commuter Times')
plt.xlabel('Time (Minutes)')
plt.show()
```
