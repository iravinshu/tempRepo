```python
import pandas as pd
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv('../dataset/titanic.csv')
```

```python
titanic_df['survival_binary'] = titanic_df['survived'].apply(lambda x: 1 if x == 1 else 0)
n = len(titanic_df)
k = titanic_df['survival_binary'].sum()
p = k / n
print(n, k, p)
```

```python
trials = 100
x = np.arange(0, trials + 1)
pmf_values = binom.pmf(x, n=trials, p=p)
cdf_values = binom.cdf(x, n=trials, p=p)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, color='purple', alpha=0.6)
plt.title('Binomial PMF - Titanic Survival')
plt.xlabel('Number of Survivors in 100 Trials')
plt.ylabel('Probability')
plt.grid(axis='y')
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, marker='o', color='orange')
plt.title('Binomial CDF - Titanic Survival')
plt.xlabel('Number of Survivors in 100 Trials')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.tight_layout()
plt.show()
```
