```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('../dataset/Mall_Customers.csv')
data
```

```python
age_data = data['Age']
age_data = age_data.sample(n=30, replace=False)
pop_avg_age = 40 # H0 Mall customers have average age of 40
```

```python
t_statics, p_value = stats.ttest_1samp(age_data, pop_avg_age)
print(f"T statistics: {t_statics:.4f}")
print(f"P value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis.")
else:
    print(f"Since the p-value {p_value:.4f} is greater than the alpha value {alpha}, we fail to reject the null hypothesis.")
```

#### For the three combined Marathon datasets, extract the top 25 records with lowest finish time. Check the null hypothesis that the average finish time of these top 25 records is less than 130minutes. test it for a = 1%, 5% and 10%

```python
import pandas as pd
marathon_results_2015 = pd.read_csv('../dataset/marathon_results_2015.csv', sep = ',')
marathon_results_2015
```

```python
import pandas as pd
marathon_results_2016 = pd.read_csv('../dataset/marathon_results_2016.csv', sep = ',')
marathon_results_2016
```

```python
import pandas as pd
marathon_results_2017 = pd.read_csv('../dataset/marathon_results_2017.csv', sep = ',')
marathon_results_2017
```

```python
df = pd.concat([marathon_results_2015, marathon_results_2016, marathon_results_2017], ignore_index=True)
df
```

```python
df['Official Time'] = pd.to_timedelta(df['Official Time']).dt.total_seconds() / 60
lowest_25 = df.nsmallest(25, 'Official Time')
lowest_25
```

```python
pop_avg_time = 130 # H0 average finish time of top 25 records is less than 130 minutes
t_statics, p_value = stats.ttest_1samp(lowest_25['Official Time'], pop_avg_time)
print(f"T statistics: {t_statics:.4f}")
print(f"P value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis.")
else:
    print(f"Since the p-value {p_value:.4f} is greater than the alpha value {alpha}, we fail to reject the null hypothesis.")
```

```python
alpha = 0.01
if p_value < alpha:
    print(f"Since the p-value {p_value:.4f} is less than the alpha value {alpha}, we reject the null hypothesis.")
else:
    print(f"Since the p-value {p_value:.4f} is greater than the alpha value {alpha}, we fail to reject the null hypothesis.")
```
