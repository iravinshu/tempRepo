```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv('../dataset/diabetes.csv')
df
```

```python
glucose_mean = df['Glucose'].mean()
glucose_std = df['Glucose'].std()
df['Glucose_zscore'] = (df['Glucose'] - glucose_mean) / glucose_std
df
```

```python
zscore_threshold = 3
outliers_glucose = df[abs(df['Glucose_zscore']) > zscore_threshold]
outliers_glucose
```

```python
print(f"Number of glucose outliers (z score > {zscore_threshold}): {len(outliers_glucose)}")
```

```python
import pandas as pd
billionaires = pd.read_csv('../dataset/billionaires.csv', sep = ',')
billionaires
```

```python
final_worth_mean = billionaires['finalWorth'].mean()
final_worth_std = billionaires['finalWorth'].std()
billionaires['finalWorth_zscore'] = (billionaires['finalWorth'] - final_worth_mean) / final_worth_std
billionaires
```

```python
outliers_final_worth = billionaires[abs(billionaires['finalWorth_zscore']) > zscore_threshold]
outliers_final_worth
```

```python
print(f"Number of final worth outliers (z score > {zscore_threshold}): {len(outliers_final_worth)}")
```
