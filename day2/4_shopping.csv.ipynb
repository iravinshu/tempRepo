```python
import pandas as pd
import numpy as np
from unicodedata import category
```

```python
numeric_cols = ['CustomerID', 'Tenure_Months', 'Transaction_ID', 'Quantity', 'Avg_Price', 'Delivery_Charges', 'GST', 'Offline_Spend', 'Online_Spend', 'Month', 'Discount_pct']
```

```python
df = pd.read_csv('shopping.csv')
```

```python
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Filter rows by a single col value

```python
male_customer = df[df['Gender'] == 'M']
male_customer
```

```python
male_customer_all = df.query("Gender == 'M'")
male_customer_all
```

### Filter rows by multiple col conditions

```python
male_chicago = df[(df['Gender'] == 'M') & (df['Location'] == 'Chicago')]
male_chicago
```

```python
male_chicago_all = df.query("Gender == 'M' & Location == 'Chicago'")
male_chicago_all
```

### Filter rows using isin()

```python
selected_products = df[df['Product_Category'].isin(['Nest-USA', 'Office'])]
selected_products
```

```python
categories = ['Nest-USA', 'Office']
selected_products_all = df.query("Product_Category in @categories")
selected_products
```

### Filter rows by numeric conditions

```python
multiple_qty = df[df['Quantity'] > 2]
multiple_qty
```

```python
multiple_qty_all = df.query("Quantity > 2")
multiple_qty_all
```

### Filter rows using between

```python
tenure_range = df[df['Tenure_Months'].between(6, 12)]
tenure_range
```

```python
print(df['Tenure_Months'].drop_duplicates().sort_values())
```

```python
print(df['Product_Category'].drop_duplicates().sort_values())
```

```python
tenure_range_all = df.query("6 <= Tenure_Months <= 12")
tenure_range_all
```

### Filter rows by date range

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
jan_2019 = df[(df['Transaction_Date'] >= '2019-01-01') & (df['Transaction_Date'] < '2019-02-01')]
jan_2019
```

### Filter string columns with starswith / endswtih - safe for NaNs

```python
electronics_products = df[df['Product_Description'].str.startswith('Elect', na=False)]
electronics_products
```

```python
electronics_products_all = df[df['Product_Description'].str.endswith('USA', na=False)]
electronics_products_all
```

### Group based filtering total spend per customer

```python
customer_total_online = df.groupby('CustomerID')['Online_Spend'].sum()
customer_total_online
```

```python
high_spenders = customer_total_online[customer_total_online > 5000]
high_spenders
```

```python
high_spenders_all = df.groupby('CustomerID').filter(lambda x: x['Online_Spend'].sum() > 5000)
high_spenders_all
```
